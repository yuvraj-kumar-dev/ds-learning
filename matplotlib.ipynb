```python
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt

# Load the diabetes dataset
dataset = datasets.load_diabetes()

# dict_keys(['data', 'target', 'frame', 'DESCR', 'feature_names', 'data_filename', 'target_filename', 'data_module'])

dataset1 = dataset.data[:, np.newaxis, 2]  # Use the third feature (index 2) as a single feature for regression

X_train, X_test, Y_train, Y_test = train_test_split(dataset1, dataset.target, test_size=0.2, random_state=42)

model = linear_model.LinearRegression()
model.fit(X_train, Y_train)
predictions = model.predict(X_test)

plt.scatter(X_test, Y_test, color='black', label='Data')
plt.plot(X_test, predictions, color='blue', linewidth=3, label='Predicted')
plt.show()

```

```python
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt


# Load the iris dataset

dataset = datasets.load_iris()

# 0 - Iris-Setosa
# 1 - Iris-Versicolour
# 2 - Iris-Virginica

# Features and labels

features = dataset.data
labels = dataset.target

# Training the classifier

clf = KNeighborsClassifier()
clf.fit(features, labels)

predictions = clf.predict([[31,1,1,1]])
print(predictions)


plt.scatter(features[:, 0], features[:, 1], c=labels, cmap='viridis', edgecolor='k', s=20)
plt.show()
```
