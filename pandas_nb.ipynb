*What is Pandas*

PANDAS is a library in python used to manipulate and analyse data. PANDAS is used to read and preprocess the data present in JSON, CSV, EXCEL files. *Example:* for example if you have a excel data of employees of a company and you want to filter out those employees who work with python and create a another excel for them. In this case we can use PANDAS to filter out our required data. PANDAS effectively handle large datasets.

# Concatenate DataFrames

```python
df4 = pd.DataFrame({"name": ["Ram", "Shyam"], "marks": [1,3]})

df4
```

```python
df5 = pd.DataFrame({"name":["Ramu", "Kaka"], "marks": [5,10]})
df5
```

```python
pd.concat([df4, df5])
```

# Merge DataFrames

```python
df6 = pd.DataFrame({"name": ["Michael", "Jackson", "Jai", "Kishan"], "marks": [1,2,34,3]})
df6
```

```python
df7 = pd.DataFrame({"name": ["Michael", "Jackson", "Jai", "Kishan"], "roll": [10,11,12,13]})
df7
```

```python
pd.merge(df6, df7, on="name")
```
